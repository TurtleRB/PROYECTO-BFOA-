```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df1 = 'resultados1.csv'  
df2 = 'resultados2.csv'  
data1 = pd.read_csv(df1)
data2 = pd.read_csv(df2)
nfe1 = data1['NFE'].iloc[0]
nfe2 = data2['NFE'].iloc[0]

# Calcular el promedio de "Mejor Fitness"
promedio1 = np.mean(data1['Mejor Fitness'])
promedio2 = np.mean(data2['Mejor Fitness'])

# Configuración de la gráfica
plt.figure(figsize=(10, 6))
plt.bar(1, promedio1, color='purple', width=0.5, label='NFE BOFA original: ' + str(nfe1))
plt.bar(2, promedio2, color='pink', width=0.5, label='NFE BOFA mejorado: ' + str(nfe2))

# Añadir etiquetas y título
plt.ylabel("Mejor Fitness")
plt.title("Comparación de Promedios de Mejor Fitness entre BFOA original y BFOA mejorado")
plt.xticks([1, 2], ['BFOA original', 'BFOA mejorado'])  # Etiquetas en el eje X
plt.legend()

# Mostrar valores encima de las barras
plt.text(1, promedio1 + 50, f'{promedio1:.2f}', ha='center', va='bottom', color='black')
plt.text(2, promedio2 + 50, f'{promedio2:.2f}', ha='center', va='bottom', color='black')

plt.show()
```
